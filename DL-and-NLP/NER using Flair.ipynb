```
from flair.data import Sentence
from flair.models import SequenceTagger

# Load tagger
tagger = SequenceTagger.load("flair/ner-english")

# Define a function to predict NER tags for a given sentence
def predict_ner_tags(sentence):
    # Create a Sentence object
    sentence_obj = Sentence(sentence)
    
    # Predict NER tags
    tagger.predict(sentence_obj)
    
    # Extract NER tags in the desired format
    ner_tags = []
    for entity in sentence_obj.get_spans('ner'):
        ner_tags.append((entity.tag, entity.text))
    
    # Print sentence
    print(f"\nProcessing sentence: {sentence}")
    
    # Print predicted NER tags in the desired format
    print('The following NER tags are found:')
    for tag, text in ner_tags:
        print(f'"{text}" → {tag}')

# Example usage
sentence = "George Washington went to Washington"
predict_ner_tags(sentence)
```
