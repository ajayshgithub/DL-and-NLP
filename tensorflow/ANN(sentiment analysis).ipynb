```
import pandas as pd
import numpy as np
import tensorflow as tf
```

## Collect data

```
df = pd.read_csv(@SYS.DATASANDBOX_PATH + '4031682022/Repo/labeledTrainData.tsv',header=0, delimiter="\t",quoting=3)
```

```
print('Number of examples: ', df.shape)
df.head()
```

There are 3 columns in the data: 
"id": movie id,
"sentiment": 0--> -ve  , 1--> +ve
"review": movie review

```
#Check out first review
df.review[0]
```

```
#Extract review and sentiments
reviews = df['review'].tolist()
sentiment = df['sentiment'].tolist()
```

## Convert text to numbers 

Build keras tokenizer


```
#Initialize Tokenizer, max vocab =5000
tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=6000)

#Let tokenizer look at all the text
tokenizer.fit_on_texts(reviews)
```

```
#Vocabulary
len(tokenizer.word_index)
```

Convert text into numbers usinf Tf-IDF

```
#convert text into numbers
features = tokenizer.texts_to_matrix(reviews, mode='tfidf')
```

```
#input data shape
features.shape
```

```
#Check numbers
features[0]
```

```
#check length of first number
len(features[0])
```

## Build the Graph

```
#Initialize model,reshape & normalization
model= tf.keras.models.Sequential()

#normalize data
model.add(tf.keras.layers.BatchNormalization(input_shape=(5000,)))

#Add Dense Layers
model.add(tf.keras.layers.Dense(200, activation='relu'))
model.add(tf.keras.layers.Dense(100, activation='relu'))
model.add(tf.keras.layers.Dense(60, activation='relu'))
model.add(tf.keras.layers.Dense(30, activation='relu'))
```

```
#Output layer
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

```
#Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

## Model Training

```
model.summary()
```

```
features_np = np.array(features)
sentiment_np = np.array(sentiment)
```

```
#Train the model
model.fit(features_np,
          sentiment_np,
          validation_split=0.3, #80% data for Training and 20% for test
          epochs=10,
          batch_size=64)
```

#### Model Prediction

Predicting on first movie review

```
#Get the first review. Please note text is already converted to numbers.
input_data = features[0:1]

#Predict
model.predict(input_data)
```

Predicting on any text

```
#Let's predict on the review below
new_movie_review = 'Very nice movie'
```

```
#1. Convert movie review text to number using the tokenizer built above
movie_review_tfidf = tokenizer.texts_to_matrix([new_movie_review], mode='tfidf')
print('Movie review data shape',movie_review_tfidf.shape)
```

```
#Predict probability of review being positive
model.predict(movie_review_tfidf)
```
